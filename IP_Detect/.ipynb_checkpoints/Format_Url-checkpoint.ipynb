```python
from itertools import izip

with open('./IP') as IP, \
    open('./Port') as Port, \
    open('./Type') as Type :
        for line1, line2, line3 in izip(IP, Port, Type) :
            key = line3.strip().lower()
            value = line3.strip().lower() + '://' + line1.strip() + ':' + line2.strip()
            format_url = {key : value}
            print str( format_url ) + ','
```
